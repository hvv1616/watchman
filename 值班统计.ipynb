```python
import pandas as pd
import xlrd
import xlsxwriter

filename = '.\值班统计表20181231.xlsx'
sheet_no = 0

df_readin = pd.read_excel(filename, sheet_name=sheet_no)
df_readin
```

```python
name_list_watch = {}
name_list_foreman = {}
year = ['2018',]

i = 0
while True :
    try:
        if str(df_readin.iloc[i,1]) in year:
            if str(df_readin.iloc[i,0]) == '值班':
                if df_readin.iloc[i,7] in name_list_watch:
                    name_list_watch[df_readin.iloc[i,7]] += 1
                else:
                    name_list_watch.update({df_readin.iloc[i,7]:1})
            else:
                if str(df_readin.iloc[i,7]) != 'nan':
                    if df_readin.iloc[i,7] in name_list_foreman:
                        name_list_foreman[df_readin.iloc[i,7]] += 1
                    else:
                        name_list_foreman.update({df_readin.iloc[i,7]:1})
    except:
        print('类型,', '姓名,', '次数')
        for name,counter in name_list_watch.items():
            print('值班,', name, ',', counter)
        for name,counter in name_list_foreman.items():
            print('带班,', name, ',', counter)
        df_watch = pd.DataFrame(name_list_watch, index=['值班'])
        df_foreman = pd.DataFrame(name_list_foreman, index=['带班'])
        df_out = df_watch.append(df_foreman)
        df_out.T.to_excel('值班表.xlsx', sheet_name='值班', index=True, engine='xlsxwriter')
        # df_foreman.T.to_excel('带班表.xlsx', sheet_name='带班', index=False, engine='xlsxwriter')
        break
    i += 1
        
```
